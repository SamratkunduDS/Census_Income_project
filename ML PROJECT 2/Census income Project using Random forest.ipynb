```python
df = df.rename(columns = {'Unnamed: 14' : 'annual_sales'})
```

```python
#as we can see there three columns that contain unnecessary information. so we can delete it.
```

```python
df.drop(df[df['workclass']== '?'].index,inplace = True)
df.drop(df[df['occupation']=='?'].index,inplace = True)
df.drop(df[df['native-country']=='?'].index,inplace = True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
# next step is to turn the object into numbers for that i am gonna use label encoder.
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
list = []
for i in df.columns:
    if ((df[i].dtypes == object) & (i!= 'annual_sales')): #since my target variable is annual sales
        list.append(i)
```

```python
list
```

```python
#lets us transform the object column now
for i in list:
    df[i] = labelencoder.fit_transform(df[i])
```

```python
#it is transformed. Now we have to check the multicollinearity among the independant variable.
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

ls = []
for i in df.columns:
    if((df[i].dtypes!= 'object')&(i != 'annual_sales')):
        ls.append(i)
```

```python
ls
```

```python
#split the data into x and y
x = df[ls]
y = df['annual_sales']
```
